# To fill all the NaN values which data type is the object datatype

```python
df=pd.read_csv("loan-Copy1.csv")
df.head(15)
```

```python
df.select_dtypes(include="object").isnull().sum()
```

```python
for i in df.select_dtypes(include="object").columns:
    df[i].fillna(df[i].mode()[0], inplace=True)# it will also fill the object type NaN value's column
```

```python
for i in df.select_dtypes(include="object").columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include="float64").isnull().sum()
```

```python
for i in df.select_dtypes(include="float64").columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.select_dtypes(include="float64").isnull().sum()
```
